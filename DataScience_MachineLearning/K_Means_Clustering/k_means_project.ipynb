# Exploratory Data Analysis

```python
# scatterplot of Grade.Rate vs Room.Board, colored by Private column
sns.lmplot('Room.Board','Grad.Rate',data=data, hue='Private',palette='coolwarm',size=6,aspect=1,fit_reg=False)
```

```python
# scatterplot of F.Undergrad vs Outstate, colored by Private
sns.lmplot('Outstate','F.Undergrad',data=data,hue='Private',palette='coolwarm',size=6,aspect=1,fit_reg=False)
```

```python
# stacked histogram showing Out of State Tuition based on Private column
sns.set_style('darkgrid')
g = sns.FacetGrid(data,hue='Private',palette='coolwarm',size=6,aspect=2)
g.map(plt.hist,'Outstate',bins=20,alpha=0.7)
```

```python
# similar histogram but with Grad.Rate
g = sns.FacetGrid(data,hue='Private',palette='coolwarm',size=6,aspect=2)
g.map(plt.hist,'Grad.Rate',bins=20,alpha=0.7)
```

# K Means Cluster Creation

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=2)
```

```python
kmeans.fit(data.drop('Private',axis=1))
```

What are clusters centers?

```python
kmeans.cluster_centers_
```

# Evalutaion

```python
def converter(cluster):
    if cluster == 'Yes':
        return 1
    else:
        return 0
```

```python
data['Cluster'] = data['Private'].apply(converter)
```

```python
data.head()
```

```python
from sklearn.metrics import confusion_matrix,classification_report
print(confusion_matrix(data['Cluster'],kmeans.labels_))
print(classification_report(data['Cluster'],kmeans.labels_))
```
